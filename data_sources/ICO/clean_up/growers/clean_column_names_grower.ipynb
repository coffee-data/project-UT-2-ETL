```python
import pandas as pd
```

```python
# read region id data into pandas
df_countries = pd.read_csv('../../../DataHub/country_codes/data.csv')
df_countries.head()
```

```python
# read grower col names into pandas
df = pd.read_csv('grow_col_names.csv')

# skip over the year column name
df = df.iloc[1:]
df.head()
```

```python
lst_id = list(df_countries['Country_Name'])
lst_original = list(df['Country'])
```

```python
for country in lst_original:
    for id in lst_id:
        if country.lower() in id.lower():  # lower() just in case
            
            print(f"""
            -----
            Changing
            
            Original: {country}
            New: {id}
            """)
            
            country = id
```
